# Question 4 - Decision Trees

Question 4: Train Decision Tree model on train data for criterions = {’gini’, ’entropy’} and report the accuracies on the validation data. Select the best criterion and report the accuracy on the test data. (1 mark)

```python
X_train = pd.read_csv('/Users/pauldefusco/Documents/DSE/pdefusco/DSE220/wine_train_data.csv')
```

```python
Y_train = pd.read_csv('/Users/pauldefusco/Documents/DSE/pdefusco/DSE220/wine_train_labels.csv')
```

```python
X_val = pd.read_csv('/Users/pauldefusco/Documents/DSE/pdefusco/DSE220/wine_val_data.csv')
Y_val = pd.read_csv('/Users/pauldefusco/Documents/DSE/pdefusco/DSE220/wine_val_labels.csv')
X_test = pd.read_csv('/Users/pauldefusco/Documents/DSE/pdefusco/DSE220/wine_val_data.csv')
Y_test = pd.read_csv('/Users/pauldefusco/Documents/DSE/pdefusco/DSE220/wine_val_labels.csv')
```

```python
#wine_test_data.csv
#wine_val_data.csv
#wine_test_labels.csv
#wine_val_labels.csv
```

```python
y_val=Y_val['class']
y_val=y_val.as_matrix()
```

```python
# Define Model with entropy criterion
clf_ent = DecisionTreeClassifier(criterion='entropy')
# Train
clf_ent.fit(X_train, Y_train)
# Predict
predictions_ent = clf_ent.predict(X_val)
print ('Accuracy = ' + str(np.sum(predictions_ent == y_val)/len(y_val)))
```

```python
# Define Model with entropy criterion
clf_gini = DecisionTreeClassifier(criterion='gini')
# Train
clf_gini.fit(X_train, Y_train)
# Predict
predictions_gini = clf_gini.predict(X_val)
print ('Accuracy = ' + str(np.sum(predictions_gini == y_val)/len(y_val)))
```

The best model criterion is entropy

```python
X_train_new = np.concatenate((X_train, X_val))
Y_train_new = np.concatenate((Y_train, Y_val))
```

```python
clf_gini_test = DecisionTreeClassifier(criterion='entropy')
clf_gini_test.fit(X_train_new, Y_train_new)
predictions_gini_test = clf_gini_test.predict(X_test)
```

```python
y_test=Y_test['class']
y_test=y_test.as_matrix()
```

```python
print ('Accuracy = ' + str(np.sum(predictions_gini_test == y_test)/len(y_test)))
```

Question 5: Use the criterion selected above to train Decision Tree model on train data for min samples split={2,5,10,20} and report the accuracies on the validation data. Select the best parameter and report the accuracy on the test data. (2 marks)

```python
for sample_split in [2,5,10,20]:
    clf = DecisionTreeClassifier(criterion='entropy', min_samples_split=sample_split,min_samples_leaf=1)
    clf.fit(X_train, Y_train)
    y_pred = clf.predict(X_val)
    print ('Validation accuracy for min_samples_split ' + str(sample_split) + ' = ' + str(np.sum(y_pred == y_val)*1.0/len(y_val)))
```

The best accuracy is obtained with a minimum sample split of 2

```python
clf = DecisionTreeClassifier(criterion='entropy', min_samples_split=5)
clf.fit(X_train_new, Y_train_new)
y_pred_test = clf.predict(X_test)
print ('Validation accuracy for min_samples_split ' + '2' + ' = ' + str(np.sum(y_pred_test == y_test)*1.0/len(y_test)))
```

Question 6: Use the parameters selected above (Q4 and Q5) to train Decision Tree model using the first 20, 40, 60, 80 and 100 samples from train data. Keep the validation set unchanged during this analysis. Report and plot the accuracies on the validation data. (2 marks)

```python
accuracy = []
i = 0
for k in [20,40,60,80,100]:
    X_train_sample = X_train.ix[i:k,]
    Y_train_sample = Y_train.ix[i:k,]
    i+=20
    clf = DecisionTreeClassifier(criterion='entropy', min_samples_split=20)
    clf.fit(X_train_sample, Y_train_sample)
    y_pred = clf.predict(X_val)
    print ('Accuracy = ' + str(np.sum(y_pred == y_val)/len(y_val)))
    acc_i = np.sum(y_pred == y_val)/len(y_val)
    accuracy.append(acc_i)
```

```python
import matplotlib.pyplot as plt
plt.plot([20,40,60,80,100],accuracy,'ro')
plt.ylabel('Accuracy')
plt.xlabel('Sample Size')
plt.show()
```

# Question 5 - Nearest Neighbor

Normalize Data: Normalize features such that for each feature the mean is 0 and the standard deviation is 1 in the train+validation data. Use the normalizing factors calculated on train+validation data to modify the values in train, validation and test data.

```python
from sklearn import preprocessing
```

Extract Mean and Standard Dev from original train + validation dataset. 

```python
X_train_new = pd.DataFrame(X_train_new, columns=X_train.columns)
means = []
stdevs = []
for k in X_train_new.columns:
    means.append(X_train_new[k].mean())
    stdevs.append(X_train_new[k].std())
```

Now rescale the train + validation dataset

```python
X_train_val_scaled = preprocessing.scale(X_train_new)
```

```python
X_train_val_scaled = pd.DataFrame(X_train_val_scaled, columns=X_train.columns)
```

Modify Train Data

```python
j=0
train=[]
for k in X_train.columns:
    for i in X_train[k]:
        new_x = (i-means[j])/stdevs[j]
        train.append(new_x)
    j+=1       
```

```python
new1 = np.asarray(train)
new2 = new1.reshape(100,13)
X_train_scaled = pd.DataFrame(new2, columns=X_train.columns)
```

Modify Validation Data

```python
j=0
val=[]
for k in X_val.columns:
    for i in X_val[k]:
        new_x = (i-means[j])/stdevs[j]
        val.append(new_x)
    j+=1       
```

```python
new1 = np.asarray(val)
new2 = new1.reshape(39,13)
X_val_scaled = pd.DataFrame(new2, columns=X_val.columns)
```

Modify Test Data

```python
j=0
test=[]
for k in X_test.columns:
    for i in X_test[k]:
        new_x = (i-means[j])/stdevs[j]
        test.append(new_x)
    j+=1       
```

```python
new1 = np.asarray(test)
new2 = new1.reshape(39,13)
X_test_scaled = pd.DataFrame(new2, columns=X_test.columns)
```

Question 9: Train the k-nn model on train data for k=1,3,5,7,9. Report and plot the accuracies on the validation data. Select the best ’k’ value and report the accuracy on the test data for the selected ’k’. Use Euclidean distance. (2 marks)

```python
acc = []
best_acc = 0.0
best_k = 1
for n in range(1,10,2): 
    clf = KNeighborsClassifier(n)
    clf.fit(X_train_scaled, Y_train)
    predictions = clf.predict(X_val_scaled)
    accuracy = np.sum(predictions == y_val)/(len(y_val))
    acc.append(accuracy)
    print ("Accuracy = " + str(accuracy) + " at k = " + str(n))
    if (accuracy > best_acc):
        best_k = n
        best_acc = accuracy
```

```python
plt.plot(range(1,10,2),acc,'ro')
plt.ylabel('Accuracy')
plt.xlabel('Choice of k')
plt.show()
```

```python
clf = KNeighborsClassifier(best_k)
clf.fit(X_train_scaled, Y_train)
predictions = clf.predict(X_test_scaled)
accuracy = np.sum(predictions == y_val)/(len(y_val))
print ("Best Accuracy = " + str(accuracy) + " at k = " + str(best_k))
```
