```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
l = []
with open('Animals_with_Attributes/predicate-matrix-continuous.txt', 'r') as f:
    for line in f:
        line = line.strip().split(',')
        if len(line)>0:
            l.append(line)
print l
```

```python
floats = []
for i in l:
    for k in i:
        new = [float(x) for x in k.split()]
        floats.append(new)
```

```python
pred_matrix = np.array(floats)
```

```python
pred_matrix
```

```python
nms = []
with open('Animals_with_Attributes/classes.txt', 'r') as f:
    for line in f:
        line=line.replace('\t',' ').replace('+', ' ')
        line = line.strip()
        if len(line)>0:
            nms.append(line)
print nms
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 10).fit_predict(pred_matrix)
```

```python
cluster_0 = []
cluster_1 = []
cluster_2 = []
cluster_3 = []
cluster_4 = []
cluster_5 = []
cluster_6 = []
cluster_7 = []
cluster_8 = []
cluster_9 = []

for i,k in zip(nms,kmeans):
    if k == 0:
        cluster_0.append(i)
    if k == 1:
        cluster_1.append(i)
    if k == 2:
        cluster_2.append(i)
    if k == 3:
        cluster_3.append(i)
    if k == 4:
        cluster_4.append(i)
    if k == 5:
        cluster_5.append(i)
    if k == 6:
        cluster_6.append(i)
    if k == 7:
        cluster_7.append(i)
    if k == 8:
        cluster_8.append(i)
    if k == 9:
        cluster_9.append(i)
```

```python
print cluster_0
print cluster_1
print cluster_2 
print cluster_3 
print cluster_4 
print cluster_5 
print cluster_6 
print cluster_7 
print cluster_8 
print cluster_9
```

The clustering kind of makes sense as similar animals belong to the same clusters. It is not perfect though...

```python
from pylab import rcParams
rcParams['figure.figsize'] = 5,10
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
lk_matrix = linkage(pred_matrix, 'ward')
```

```python
plt.title('Dendrogram')
plt.xlabel('Distance')
plt.ylabel('Animals')
dendrogram(lk_matrix, orientation='right' ,labels=nms)
plt.show()
```

Hierarchical clustering seems to mostly have made sense in grouping the animals based on similarity (we have all sea animals in one tree leaf). It does a better job than the initial attempt because it provides additional insights into animal similarity by reflecting additional dimensions

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
twod = pca.fit_transform(pred_matrix)
```

```python
animals = [i[2:] for i in nms]
```

```python
rcParams['figure.figsize'] = 10,10
fig = plt.figure()
ax = fig.add_subplot(111)
for i, k in zip(animals, twod):
    plt.plot(twod, "o")
    ax.annotate(i,k)
```
