```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state = 1)
```

```python
clf = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
print("Training R2: ", clf.score(X_train, y_train))
print("Test R2: ", clf.score(X_test, y_test))
```

```python
import graphviz
from sklearn import tree
```

```python
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph
```

Same tree with Max Depth parameter - higher depth, higher complexity

```python
clf = DecisionTreeClassifier(max_depth = 3).fit(X_train, y_train)
```

```python
print("Training R2: ", clf.score(X_train, y_train))
print("Test R2: ", clf.score(X_test, y_test))
```

```python
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph
```

Seems to overfit a bit less... other parameters to play with are minsamplesleaf and maxleafnodes
