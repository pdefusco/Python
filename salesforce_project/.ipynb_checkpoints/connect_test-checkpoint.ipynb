```python
contact_owners = []
phone_numbers = []
profiles = []
mobile_numbers = []
other_phones = []
fax_numbers = []
reports_to = []
assistants = []
lead_sources = []
asst_phones = []
other_address = []

for i in range(100):
    contact_owners.append(fake.name())
    phone_numbers.append(fake.phone_number())
    profiles.append(fake.profile())
    mobile_numbers.append(fake.phone_number())
    other_phones.append(fake.phone_number())
    fax_numbers.append(fake.phone_number())
    reports_to.append(fake.phone_number())
    assistants.append(fake.name())
    lead_sources.append(fake.name())
    asst_phones.append(fake.phone_number())
    other_address.append(fake.address())
```

```python
contacts = []
contacts = pd.DataFrame(profiles)
```

```python
contacts['Contact Owner'] = contact_owners
contacts['Phone'] = phone_numbers
contacts['MobilePhone'] = mobile_numbers
contacts['Other Phone'] = other_phones
contacts['Fax'] = fax_numbers
contacts['Reports To'] = reports_to
contacts['AssistantName'] = assistants
contacts['Lead Source'] = lead_sources
contacts['Asst Phone'] = asst_phones
contacts['Other Address'] = other_address
```

```python
contacts.columns
```

```python
firstnames = []
lastnames = []
for i in contacts.name:
    firstnames.append(i.split(' ')[0])
    lastnames.append(i.split(' ')[1])
```

```python
contacts = contacts.drop(columns=['company','Other Address','address','Contact Owner',
                                  'website','blood_group', 'sex', 'ssn', 'username', 
                                  'current_location','residence','Reports To', 'name'])
```

```python
len(contacts.columns)
```

```python
contacts = contacts.rename(columns={
                         #'address':'MailingAddress', 
                         'birthdate':'Birthdate', 
                         #'company':'AccountId', 
                         'job':'Title',
                         'mail':'Email', 
                         'name':'Name',
                         'Asst Phone':'AssistantPhone', 
                         #'Reports To':'ReportsToId', 
                         #'Other Address':'OtherAddress',
                         'Lead Source':'LeadSource', 
                         'Other Phone':'OtherPhone', 
                         'Phone Number':'PhoneNumber', 
                         'Contact Owner':'ContactOwner'
                        })
```

```python
len(firstnames)
```

```python
contacts['Birthdate'] = [str(i) for i in contacts['Birthdate']]
contacts['FirstName'] = firstnames
contacts['LastName'] = lastnames
```

```python
contacts.to_dict('records')
```

```python
for i in contacts.to_dict('records'):
    sf.Contact.create(i)
```
