```python
df = pd.read_pickle(dirname+'final_df_pickle.pickle')
```

```python
#Creating day of the week feature
df['weekday'] = df.Date.dt.weekday
```

```python
#Creating total number of days in month
lst = []
for i,k in zip(df['year'], df['month']):
    lst.append(calendar.monthrange(i, k)[1])
df['month_total_days'] = lst
```

```python
#Creating is holiday flag
df['is_holiday'] = [1 if i in holidays.Portugal() else 0 for i in df['Date']]
```

```python
#Also adding holiday names
hol_names = []
for i in df['Date']:
    if i in holidays.Portugal():
        hol_names.append(holidays.Portugal().get(i))
    else:
        hol_names.append("No Holiday")
df['holiday_name'] = hol_names
```

```python
#According to research, most government employees are paid ont he 20th 
#while private industry employees are paid on the 30th
df['biz_payday'] = [1 if i == 20 else 0 for i in df['day']]
df['govt_payday'] = [1 if i == 30 else 1 if i == 29 else 1 if i == 28 else 0 for i in df['day']]
```
