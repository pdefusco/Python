```python
def pointb_corr(df, y):
    #Correlation between independent features and (binary) target var
    pbcorr_list = []
    for i in df.columns:
        if math.isnan(stats.pointbiserialr(df[i],y)[0]) != True:
            pbcorr_list.append((str(i), stats.pointbiserialr(df[i],y)[0]))
    return pbcorr_list
```

```python
#Function to create a pipeline with preprocessing steps and a given classifier
#CLF is an instance of Sklearn Estimator

def create_pipeline(clf):
    
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', RobustScaler())])

    preprocessor = ColumnTransformer(
        transformers=[('num', numeric_transformer, numeric_features)], remainder='passthrough')
    
    pca = PCA()
    clf_name = type(clf).__name__

    # Append classifier to preprocessing pipeline.
    pipe = Pipeline(steps=[('preprocessor', preprocessor),('pca', pca),(clf_name, clf)])

    return pipe
```

```python
#Split data between training and test sets
def split_data(df, target_name, test_size):
    
    y = df[target_name]
    X = df.drop(target_name, axis=1)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state = 1)

    return X_train, X_test, y_train, y_test
```

```python
#Function to execute pipeline for multiple models
#Pass a dictionary with models as keys and their respective parameters as values

def test_model(clfs, df, y_name, scoring_metric):
    
    X_train, X_test, y_train, y_test = split_data(df, y_name,0.2)
    
    cv_results = {}
    test_results = {}
    best_estimators = {}
    
    #Iterate through each model:
    for clf in clfs.items():
        
        pipe = create_pipeline(clf[0])
    
        gs = GridSearchCV(pipe, param_grid=clf[1], iid=False, cv=5, scoring=scoring_metric, n_jobs=-1)
        
        gs.fit(X_train,y_train)
        
        cv_metric = cross_val_score(gs, X_train, y_train, cv=3)
        
        test_score = gs.score(X_test, y_test)
        
        best_estimators[type(clf[0]).__name__] = gs.best_estimator_
        
        cv_results[type(clf[0]).__name__] = cv_metric
        
        test_results[type(clf[0]).__name__] = test_score
    
    return cv_results, test_results, best_estimators
```

```python
clfs_pca = {
        AdaBoostClassifier(DecisionTreeClassifier(max_depth = 5, max_features=10, class_weight='balanced'))
                                            :{'pca__n_components':[10],
                                            'AdaBoostClassifier__learning_rate':[0.6, 0.9],
                                            'AdaBoostClassifier__n_estimators':[10]},
        
        RandomForestClassifier():{'pca__n_components':[10, 20],
                                            'RandomForestClassifier__max_depth':[10], 
                                            'RandomForestClassifier__min_samples_leaf':[20000],
                                            'RandomForestClassifier__min_samples_split':[10000], 
                                            'RandomForestClassifier__class_weight':['balanced'], 
                                            'RandomForestClassifier__n_estimators':[10]}
        }
```

```python
cv_results_pca, test_results_pca, best_estimators_pca = test_model(clfs_pca, df, 'isdefault', 'accuracy')
```

```python
#Let's try the same pipeline but use feature selection techniques rather than PCA:
def create_pipeline(clf):
    
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', RobustScaler())])
    
    #Note: because we are no longer using PCA here, it would be correct to include
    #a tranasformer for categorical features so not to lose them

    preprocessor = ColumnTransformer(
        transformers=[('num', numeric_transformer, numeric_features)], remainder='passthrough')
    
    
    var_selector = VarianceThreshold(0.5)
    kbest_selector = SelectKBest(mutual_info_classif)
    clf_name = type(clf).__name__

    # Append classifier to preprocessing pipeline.
    pipe = Pipeline(steps=[('preprocessor', preprocessor),('var', var_selector),
                           ('kbest', kbest_selector),(clf_name, clf)])

    return pipe
```

```python
def conf_matrix(clf, X_train, X_test, y_train, y_test):
    fig, ax = plt.subplots(figsize=(8,5))

    cm = ConfusionMatrix(clf, classes=[0,1], ax=ax)
    cm.fit(X_train, y_train)
    cm.score(X_test, y_test)

    ax.set(title='GBM Confusion Matrix', xlabel='Predicted Class', ylabel='True Class')

    cm.poof()
```

```python
def val_curves(clf, param_name, param_range, cv, X_train, y_train, scoring):

    fig, ax = plt.subplots(figsize=(8,4))

    viz_vc_bg = ValidationCurve(
                    clf, param_name=param_name,
                    param_range=param_range, cv=cv, 
                    scoring=scoring, ax=ax)

    viz_vc_bg.fit(X_train, y_train)
    
    clf_name = type(clf).__name__
    title = "Validation Curves for: {}".format(clf_name)
    
    ax.set(title=title, ylabel='Score')
    ax.legend().set_visible(True)

    viz_vc_bg.poof()
```

```python
def learn_curves(clf, cv, X_train, y_train, scoring):

    fig, ax = plt.subplots(figsize=(8,4))

    viz_lc_bg = LearningCurve(clf, train_sizes = np.linspace(0.1, 1.0, 10), 
                              scoring=scoring, cv=cv, ax=ax)

    # Fit and poof the visualizer
    viz_lc_bg.fit(X_train, y_train)

    ax.set(title='Learning Curves', xlabel='Training Instances', ylabel='Score')

    ax.legend().set_visible(True)

    viz_lc_bg.poof()
```

```python
val_curves(best_estimators_recall['RandomForestClassifier'], 
           'RandomForestClassifier__n_estimators', [10, 30, 60], 3, X_train, y_train, 'recall')
```

Unfortunately our recall is pretty low. Let's look at how changing the decision threshold might impact recall:

```python
y_scores = best_estimators_recall['RandomForestClassifier'].predict_proba(X_test)[:, 1]
```

```python
p, r, thresholds = precision_recall_curve(y_test, y_scores)
```

```python
def adjusted_classes(y_scores, t):

    return [1 if y >= t else 0 for y in y_scores]

def precision_recall_threshold(p, r, thresholds, t=0.5):

    y_pred_adj = adjusted_classes(y_scores, t)
    print(pd.DataFrame(confusion_matrix(y_test, y_pred_adj),
                       columns=['pred_neg', 'pred_pos'], 
                       index=['neg', 'pos']))
    
    # plot the curve
    plt.figure(figsize=(8,8))
    plt.title("Precision and Recall curve ^ = current threshold")
    plt.step(r, p, color='b', alpha=0.2,
             where='post')
    plt.fill_between(r, p, step='post', alpha=0.2,
                     color='b')
    plt.ylim([0.5, 1.01]);
    plt.xlim([0.5, 1.01]);
    plt.xlabel('Recall');
    plt.ylabel('Precision');
    
    close_default_clf = np.argmin(np.abs(thresholds - t))
    plt.plot(r[close_default_clf], p[close_default_clf], '^', c='k',
            markersize=15)
```

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    """
    Modified from:
    Hands-On Machine learning with Scikit-Learn
    and TensorFlow; p.89
    """
    plt.figure(figsize=(8, 8))
    plt.title("Precision and Recall Scores as a function of the decision threshold")
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
    plt.ylabel("Score")
    plt.xlabel("Decision Threshold")
    plt.legend(loc='best')
```

```python
plot_precision_recall_vs_threshold(p, r, thresholds)
```
