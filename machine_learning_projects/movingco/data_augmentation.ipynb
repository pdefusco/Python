```python
import pandas as pd
import numpy as np
```

```python
temp = list()
for i in zip_text.splitlines()[2:]:
    temp.append(i.split('\t'))
```

```python
temp_df = pd.DataFrame(temp)
temp_df.head()
```

```python
temp_cols = []
for i in zip_text.splitlines()[0:2]:
    temp_cols.append(i.split('\t'))
```

```python
temp_cols = [item for sublist in temp_cols for item in sublist]
```

```python
temp_cols.pop(0)
```

```python
temp_df.columns = temp_cols
```

```python
temp_df.drop(columns = ['#', 'City'], inplace = True)
```

```python
#Splitting Location column into two (Latitude and Longitude)
location_df = pd.DataFrame(list(map(lambda x: x.split(', '), temp_df.Location)), columns = ['Latitude', 'Longitude'])
zip_wealth = pd.concat([location_df, temp_df], join='outer', axis = 1)
zip_wealth.drop(columns = ['Location'], inplace = True)
```

```python
#Simplifying column name
zip_wealth.rename(columns = {'Avg. Income/H/hold':'Avg Income'}, inplace = True)
```

```python
# Cleaning up - removing symbols
zip_wealth['Avg Income'] = list(map(lambda x: x[:-3].strip('$').replace(',', ''), zip_wealth['Avg Income']))
zip_wealth['National Rank'] = list(map(lambda x: x.strip('#').replace(',',''), zip_wealth['National Rank']))
```

```python
zip_wealth['Population'] = list(map(lambda x: x.replace(',',''), zip_wealth['Population']))
```
