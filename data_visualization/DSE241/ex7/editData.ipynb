Adding Features to the Datasets Provided

```python
import pandas as pd
```

```python
schedule = pd.read_csv('Republican_primary_schedule.csv')
```

```python
lst = []
for i in schedule['state']:
    i = i.replace("_"," ")
    lst.append(i)
```

```python
schedule['state'] = lst
```

```python
timezones = pd.read_csv('timezones.csv')
```

```python
new = schedule.merge(timezones, how='left', left_on='state', right_on='State Name')
```

```python
#Samoa Standard Time
#guam - chamorro
#district eastern
```

```python
new = new.drop(columns=['State Name'])
```

```python
new = new.rename(columns={'Time Zone(s)': 'timezone'})
```

```python
new.loc[new.state=='American Samoa','timezone'] = 'Samoa Standard Time'
new.loc[new.state=='District of Columbia','timezone'] = 'Eastern Time Zone'
new.loc[new.state=='Guam','timezone'] = 'Chomorro'
new.loc[new.state=='Northern Marianas','timezone'] = 'Chomorro'
new.loc[new.state=='Virgin Islands','timezone'] = 'Atlantic Standard Time'
new.loc[new.state=='Puerto Rico','timezone'] = 'Atlantic Standard Time'
```

```python
lst = []
for i in new['timezone']:
    string = i.split(',')
    if len(string)>1:
        string.pop(1)
    lst.append(string)
flat_list = [item for sublist in lst for item in sublist]
```

```python
new['timezone'] = flat_list
```

```python
lst = []
for i, k in zip(new['month'],new['day']):
    if i == 'March':
        m = '3'
    elif i == 'April':
        m = '4'
    elif i == 'May':
        m = '5'
    elif i == 'June':
        m = '6'
    elif i == 'February':
        m = '2'
    dat = str(m) + '-' + str(k) + '-' + '2017'
    lst.append(dat)
```

```python
new['date'] = lst
```

```python
new['date'] = pd.to_datetime(new['date'])
```

```python
new.to_csv('newschedule.csv')
```
