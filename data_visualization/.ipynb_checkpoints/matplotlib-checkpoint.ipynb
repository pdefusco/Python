```python
#Top 5 Countries by immigrants:
list(df.sort_values(by=['Total'], ascending=False).iloc[0:5].index)
```

```python
#Alternative way to do it:
df.nlargest(5, 'Total').index
```

```python
top5 = df.nlargest(5, 'Total')
top5 = top5[years]
top5.columns = top5.columns.map(int)
top5.T.plot(kind='line')
```

#### Area Plots

```python
len(years)
```

```python
top5 = top5.T
```

```python
top5.index = top5.index.map(int)
```

```python
top5.plot(kind = 'area', stacked = False, alpha = 0.45, figsize = (20,10))
plt.title("Immigration Trend of Top 5 Countries")
plt.ylabel("Number of Immigrants")
plt.xlabel("Year")

plt.show()
```

#### Now Same as Above but using the Matplotlib Artist Layer 

```python
ax = top5.plot(kind = 'area', alpha = 0.35, stacked=False, figsize = (20,10))

ax.set_title("Immigration Trend of Top 5 Countries")
ax.set_ylabel("Number of Immigrants")
ax.set_xlabel("Years")

plt.show()
```

We can build the same exact area plot with stacked lines:

```python
ax = top5.plot(kind = 'area', alpha = 0.35, stacked=True, figsize = (20,10))

ax.set_title("Immigration Trend of Top 5 Countries")
ax.set_ylabel("Number of Immigrants")
ax.set_xlabel("Years")

plt.show()
```

#### Histograms

```python
count, bin_edges = np.histogram(df['2013'])
```

```python
print(count)
print(bin_edges)
```

```python
df['2013'].plot(kind = 'hist', figsize=(10,5), xticks = bin_edges)

plt.title("Histogram of Immigration to Canada in 2013")
plt.xlabel("Incoming Immigrant Origin")
plt.ylabel("Number of Immigrants")

plt.show()
```

```python
df.loc[['Denmark', 'Sweden', 'Italy'], years]
```

```python
df_temp = df.loc[['Denmark', 'Sweden', 'Norway'], years].T
```

```python
df_temp.head()
```

```python
count, bin_edges = np.histogram(df_temp, 15)
```

```python
df_temp.plot(kind = 'hist', figsize = (10,6), bins = 15, alpha = 0.6, xticks = bin_edges, 
            color = ['coral', 'darkslateblue', 'mediumseagreen'])

plt.title("Immigration from Denmark, Sweden, and Italy")
plt.xlabel("Immigrants")
plt.ylabel("Years")

plt.show()
```

```python
#Matplotlib colors:
for name, hex in mpl.colors.cnames.items():
    print(name, hex)
```

```python
#Unstacked Bins:

count, bin_edges = np.histogram(df_temp)
xmin = bin_edges[0] - 10
xmax = bin_edges[-1] + 10

#stacked histogram

df_temp.plot(kind = 'hist', figsize = (20,10), alpha = 1, bins = 15, xticks = bin_edges, 
            color = ['coral', 'bisque', 'chocolate'], stacked = False, xlim = (xmin, xmax))

plt.title("Immigration from Denmark, Sweden and Italy")
plt.xlabel("Immigrants")
plt.ylabel("Years")

plt.show()
```

#### Horizontal Bars

```python
df_top15 = df.nlargest(15, "Total")
```

```python
df_top15 = df_top15[years]
```

```python
df_top15 = df_top15.sum(axis=1)
```

```python
df_top15
```

```python
df_top15.plot(kind = 'barh', 
             color = '#A52A2A',
             figsize = (20,10))
plt.xlabel("Number of Immigrants")
plt.ylabel("Country")

for ix, val in enumerate(df_top15):
    label = (format(int(val), ',')) 
    plt.annotate(label, xy = (val, ix), color = '#A52A2A')

plt.show()
```
