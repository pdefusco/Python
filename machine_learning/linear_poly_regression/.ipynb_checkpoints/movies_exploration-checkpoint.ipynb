Let's begin some basic exploration on Numeric features

```python
temp = movies.select_dtypes(include=['int64', 'float64']).describe().round(0)
movies.select_dtypes(include=['int64', 'float64']).describe().round(0)
```

```python
import seaborn as sns
```

```python
fig, axes = plt.subplots(4,2, figsize=(10,10))

for i,col in zip(axes.flat, temp.columns):
    sns.distplot(temp[col], ax=i).set_title('Distribution for %s'%(col))
    #plt.title('Distribution for %s'%(i))
    i.set_xlabel('')

fig.suptitle('Distribution Plots for NUM Features', fontsize=18)

plt.delaxes(axes[3][1])
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
