```python
import pandas as pd
import numpy as np
```

```python
listings = pd.read_csv('listings.csv')
```

```python
listings.columns
```

```python
#just using a subset of the dataset:
listings = listings[['id','summary','city']]
```

```python
listings.head()
```

```python
#boolean for nulls
pd.isnull(listings).head()
```

```python
#substitute certain values with null with read_csv command parameter na_values:
pd.read_csv('listings.csv',na_values=['?',-11111]).head()
```

```python
#directly assign the index with read_csv command:
indx_listings = pd.read_csv('listings.csv', index_col='id')
```

```python
indx_listings.head()
```

```python
#create a unique index from two columns:
listings['new_id'] = listings['id'].astype(str) + listings['city']
```

```python
#check with boolean whether the new listings id is unique:
listings['new_id'].is_unique
```

```python
listings['new_id'].value_counts().head()
```

```python
listings.head()
```

```python
#we create a new index from the new_id field
listings.index = listings.new_id
```

```python
#select based on index:
listings.loc['11204286San Diego']
```

```python
#let's reorder the index:
listings.reindex(listings.index[::-1]).head()
```

```python
#dropping row based on its index
#notice this is not inplace
listings.drop(['13452448Solana Beach']).head()
```

```python
#dropping columns:
listings.drop(['summary'], axis=1).head()
```

```python
#slicing based on index labels (not numeric values):
listings['11204286San Diego':'13124681Chula Vista']
```

```python
#reassigning values based on index labels:
listings['11204286San Diego':'13124681Chula Vista']= 5
```

```python
listings['11204286San Diego':'13124681Chula Vista']
```

```python
#selecting columns:
listings[['id','summary']].head()
```

```python
#selecting based on column values:
listings[listings.id==5]
```

```python
listings[listings.id>5].head()
```

```python
#the query method does something analogous:
listings.query('id>5').head()
```

```python
#reference namespace variables with @ prefix:
sample_var = 100
listings.query('id<@sample_var')
```

```python
#select columns and rows based on lists:
listings.loc['11204286San Diego',['id','summary']]
```

```python
#or:
listings.loc[:'11204286San Diego',['id','summary']]
```

```python
#using iloc we can select columns and rows based on index position:
#selecting first ten rows, all columns
listings.iloc[:10,:]
```

```python
#isin method is useful to select based on membership to other list, series or df:
sample_list = ['San Diego']
listings.city.isin(sample_list).head()
bool_city = listings.city.isin(sample_list)
```

```python
#the boolean list can be used to select the df:
listings.city[bool_city].head()
```

```python
#in effect we are slicing based on index 
listings[bool_city].head()
```

```python
#alternatively one can achieve the same as above with loc and iloc:
df_sd = listings.loc[listings.city=='San Diego', ['id','summary', 'city', 'new_id']]
```

```python
df_sd.index = listings.id[listings.city=='San Diego']
```

```python
df_sd.head()
```

```python
#selection can also be done based on  null not null:
#notnull will only return not null values
df_sd[df_sd.notnull()].head()
```

```python
#is null will return boolean values
df_sd[df_sd.isnull()].head()
```

```python
#the add method allows to set a value when two series 
#comparisons are made and in some instances thec cases do not overlap:
sample_df = pd.DataFrame([i for i in range(len(df_sd.index))])
```

```python
sample_df.columns = ['col1']
```

```python
#setting sample_df instances to null just to demo add command:
sample_df.iloc[:100,:] = np.nan
```

```python
sample_df.index = df_sd.index
```
