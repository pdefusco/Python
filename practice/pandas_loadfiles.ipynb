```python
import pandas as pd
import numpy as np
```

Load CSV

```python
#simples way to load csv file:
l1 = pd.read_csv('listings.csv')
```

Read CSV important arguments:

```python
#header:
#identify which row to use as file header:
#the default is 0
l_header = pd.read_csv('listings.csv', header=2)
```

```python
l_header.head(2)
```

```python
#index_col: column to use as the row labels of the dataframe:
#let's try using scrape_id:
pd.read_csv('listings.csv', index_col='scrape_id').head()
```

```python
#if  the index_col command is not specified, 
#the index col will default to a lit of integers:
#this is equivalent to passing index_col=None
pd.read_csv('listings.csv', index_col=None).head()
```

```python
#skip rows: 
#list of rows to be skipped
#if passing an integer rather than a list, it will skip all rows up to integer index passed
pd.read_csv('listings.csv', skiprows=[2]).head()
```

```python
#na values shows values to be passed as null values:
#we pass the value for column id at row 3 and notice it is loaded as null:
pd.read_csv('listings.csv', na_values=[13124681]).head()
```

```python
#na_filter arg: 
#Detect missing value markers (empty strings and the value of na_values). 
#In data without any NAs, passing na_filter=False can improve the performance of reading a large file
pd.read_csv('listings.csv', na_filter=False).head()
```

Pickle Files:

```python
#we will first create a pickle file from a dictionary, then demo how to load it:
p_dict = {'key1':[i for i in range(10)], 
         'key2': [i for i in range(0,20,2)]}
```

```python
import pickle
```

```python
pickle.dump(p_dict, open("sample_pickle.p","wb"))
```

```python
from_pickle = pickle.load(open("sample_pickle.p", "rb"))
```

```python
from_pickle
```

```python
#pandas supports to_pickle to create a pickle file directly from a df:
listings = pd.read_csv('listings.csv')
```

```python
listings.to_pickle('listings.p')
```

```python
listings_from_pickle = pickle.load(open("listings.p", "rb"))
```

```python
#the above is equivalent to the read_pickle pandas command:
pd.read_pickle('listings.p').head()
```
