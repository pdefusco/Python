More practice on sorting and ranking

```python
import pandas as pd
import numpy as np
```

```python
listings = pd.read_csv('listings.csv')
```

```python
listings.set_index('id', inplace=True)
```

```python
listings.head(3)
```

```python
listings.sort_index().head(3)
```

```python
listings.sort_index(ascending=False).head(3)
```

```python
#sorting on the columns:
listings.sort_index(axis=1).head()
```

```python
listings.sort_index(axis=1, ascending=False).head()
```

```python
#sorting by values:
listings.index.sort_values()
```

```python
listings.zipcode.sort_values().head()
```

```python
listings.zipcode.sort_values(ascending=False).head()
```

```python
listings.dtypes
```

```python
#sorting on more than one column:
listings[['number_of_reviews', 
          'review_scores_rating', 
          'review_scores_accuracy']].sort_values(ascending=[False,True],
                                                by=['number_of_reviews',
                                                    'review_scores_rating']).head()
```

```python
#the rank method returns an index that rearranges values relative to the series:
listings.number_of_reviews.rank()
```

```python
listings[['number_of_reviews']].rank(axis=1).head()
```

```python
#notice ties are handled by default by setting them to the lowest rank possible
pd.Series([1,1,1,4,5,6]).rank()
```

```python
#however you can pass the method argument to decide how to handle ties
pd.Series([1,1,1,4,5,6]).rank(method='min')
```

```python
#however you can pass the method argument to decide how to handle ties
pd.Series([1,1,1,4,5,6]).rank(method='first')
```

```python
#notice we changed the series elements this time:
pd.Series([1,2,2,4,4,1,1]).rank(method='max')
```

```python
#Calling the DataFrame's rank method results in the ranks of all columns:
listings[['number_of_reviews', 
          'review_scores_rating', 
          'review_scores_accuracy']].rank(ascending=False).head()
```
