```python
N = 101
```

```python
for i in range(2,N):
    p = True
    for k in range(2,N):
        if(i%k==0):
            p = False
        if p:
            print(i)
```

```python
for num in range(1,N):
    if all(num%i!=0 for i in range(2,num)):
        print(num)
```

```python
for i in range(1,N):
    count = 0
    for k in range(2, (i//2)):
        if(i % k == 0):
            count = count + 1
            break
    if (count == 0 and i !=1):
        print(" %d" %i, end=' ')
```

```python
for num in range(2,N):
    if all(num%i!=0 for i in range(2,num)):
        print(num)
```

```python
import math
def is_prime(n):
    for i in range(2, int(math.sqrt(n))+1):
        if n%i == 0:
            return False
        
    return True
        
```

```python
for num in range(2,N):
    if is_prime(num):
        print(num)
```
