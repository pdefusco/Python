# Exploration of Listings Summary File

NB: This exploration notebook focuses on the listings summary file. This is important because so far we have explored the master file, which includes time series data and therefore overrepresents listings that have higher availability on the market (more time series data points).

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
list_summ = pd.read_csv('inside_airbnb/listings_summ.csv')
```

```python
list_summ = list_summ.rename(columns={"id": "listing_id"})
```

```python
list_summ.dtypes
```

```python
list_summ.describe()
```

```python
list_summ['neighbourhood'].unique()
```

```python
L = []
for w in list_summ['neighbourhood'].unique():
    L.append((w, list_summ[list_summ['neighbourhood'] == w]['price'].mean()))
    
L = sorted(L, key = lambda x: x[1], reverse = True)
```

Plot Mean Price for Each Neighbourhood

```python
df = pd.DataFrame(L, columns=['neighbourhood', 'nbghrd_mean_price'])
```

```python
plt.figure()
df.plot(x='neighbourhood',y='nbghrd_mean_price', kind='barh', figsize=(20,20))
plt.show()
```

```python
merged = list_summ.merge(df, how='left', left_on='neighbourhood', right_on='neighbourhood')
```

```python
merged.head(3)
```

```python
nghbrd_list_count = merged[['neighbourhood','listing_id']].groupby(['neighbourhood'])
```

```python
plt.figure()
nghbrd_list_count.count().sort_values(by='listing_id').unstack().plot(kind='barh', figsize=(20,20))
plt.show()
```

```python
nghbrd_roomtype_count = merged[['room_type','listing_id']].groupby(['room_type'])
```

```python
plt.figure()
nghbrd_roomtype_count.count().sort_values(by='listing_id').unstack().plot(kind='bar', figsize=(8,8))
plt.show()
```

```python
by_numreviews = merged[['neighbourhood','number_of_reviews']].groupby('neighbourhood').count()
```

```python
#Needs to be improved
plt.figure()
by_numreviews.sort_values(by='number_of_reviews').plot(kind='barh', figsize=(30,20))
plt.show()
```

```python
nghbrd_roomtype_nbghrd_count = merged[['room_type','neighbourhood','listing_id']].groupby(['room_type','neighbourhood'])
```

```python
len(list_summ['neighbourhood'].unique())
```

```python
COL_NUM=10
ROW_NUM=10
forplot = nghbrd_roomtype_nbghrd_count.size().unstack()
```

```python
fig, axes = plt.subplots(COL_NUM,ROW_NUM, 
                     sharex=True, sharey=True, figsize=(30,30))
for i, nghbrd in enumerate(forplot):
    ax = axes[i/COL_NUM, i%ROW_NUM]
    forplot[nghbrd].plot(ax=ax, kind='bar')
    ax.set_title(nghbrd)
plt.show()
```
