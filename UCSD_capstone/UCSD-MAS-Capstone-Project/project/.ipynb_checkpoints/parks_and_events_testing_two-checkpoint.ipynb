```python
import json
from StringIO import StringIO
import pandas as pd
```

```python
a = pd.ExcelFile('listings_events_information_two.xlsx')
```

```python
sheet_dict = {}

for i in range(len(a.sheet_names)):
    sheet_dict[i] = a.parse(a.sheet_names[i])
```

```python
sheet_dict
```

```python
df = sheet_dict[0]

for i in range(1, len(sheet_dict)):
    df = df.merge(sheet_dict[i], how = 'outer', left_on = 'listing_id', right_on = 'listing_id')
```

```python
df = df.fillna(0)
```

```python
for w in df.columns:
    if 'within' in w:
        df = df.rename(columns = {w: 'events_' + w})
```

```python
df
```

```python
df.min()
```

```python
df.to_csv('listings_events_information_two.csv')
```
