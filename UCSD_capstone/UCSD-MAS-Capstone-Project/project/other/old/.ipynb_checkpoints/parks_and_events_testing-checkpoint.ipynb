```python
from collections import defaultdict
col_counts = defaultdict(int)
col_ix = df.first_valid_index()
```

```python
cols = []
for col in df.ix[col_ix].index:
    cnt = col_counts[col]
    col_counts[col] += 1
    suf = '_' + str(cnt) if cnt else ''
    cols.append(col + suf)
```

```python
df.columns = cols
df = df.drop([col_ix])
```

```python
df = df.T.drop_duplicates().T
```
