# Vector Autoregression

This notebook performs Autoregression with multiple dependent variables (the buildSegments notebook does a univariate regression)

```python
import statsmodels.api as sm
from statsmodels.tsa.api import VAR, DynamicVAR
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
%matplotlib inline
```

```python
master = pd.read_parquet('master.parquet',engine='fastparquet')
```

```python
master.select_dtypes(include=['int64','float64']).columns
```

```python
master_subset = master[['price_y', 'transit_length', 'beds','review_scores_rating']]
```

```python
master_subset.index = master['date']
```

```python
master_subset.index = pd.DatetimeIndex(master_subset.index,start='0', end='lngth')
```

Note The VAR class assumes that the passed time series are stationary. Non-stationary or trending data can often be transformed to be stationary by first-differencing or some other method. For direct analysis of non-stationary time series, a standard stable VAR(p) model is not appropriate.

```python
master_subset.head(3)
```

```python
master_subset.isnull().sum()
```

```python
master_subset = master_subset.dropna()
```

```python
model = VAR(master_subset)
```

```python
results = model.fit(2)
```

```python
results.summary()
```

```python
plt.figure(figsize=(20,20))
results.plot()
plt.show()
```

```python
results.plot_acorr()
```

Lag order selection

```python
model.select_order(15)
```
