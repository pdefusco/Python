```python
nghbrd_roomtype_nbghrd_count = listings_other[['bed_type','neighbourhood','id']].groupby(['bed_type','neighbourhood'])
```

```python
listings_other['neighbourhood'].nunique()
```

```python
COL_NUM=8
ROW_NUM=13
forplot = nghbrd_roomtype_nbghrd_count.size().unstack()
```

```python
fig, axes = plt.subplots(COL_NUM,ROW_NUM, 
                     sharex=True, sharey=True, figsize=(30,30))
for i, nghbrd in enumerate(forplot):
    ax = axes[i/COL_NUM, i%ROW_NUM]
    forplot[nghbrd].plot(ax=ax, kind='bar')
    ax.set_title(nghbrd)
plt.show()
```
