```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
```

```python
decile = lambda grade: grade / 10 * 10
```

```python
from collections import Counter 
histogram = Counter(decile(grade) for grade in grades)
```

```python
plt.bar([x-4 for x in histogram.keys()],histogram.values(),8)
plt.axis([-5,105,0,5])
```

```python
plt.xticks([10 * i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('# of Students')
plt.title('Distribution of Exam 1 Grades')
plt.show()
```

```python
variance = [1]
for k in range(1,9):
    i = variance[k-1]*2
    variance.append(i)
```

```python
bias_squared = []
for i in reversed(variance):
    bias_squared.append(i)
```

```python
total_error = [x+y for x,y in zip(variance,bias_squared)]
```

```python
xs = [i for i, _ in enumerate(variance)]
```

```python
plt.plot(xs,variance, 'g-', label='variance')
plt.plot(xs,bias_squared, 'r-.', label='bias^2')
plt.plot(xs,total_error, 'b:', label='total error')
```

```python
plt.legend(loc=9)
plt.xlabel('model complexity')
plt.title('The Bias-Variance Tradeoff')
plt.show()
```

```python
import random
import string
friends = random.sample(range(70, 90), 15)
minutes = random.sample(range(100, 200), 15)
labels = list(string.ascii_lowercase[:15])
```

```python
plt.scatter(friends,minutes)
```

```python
for label, friend_ct, minute_ct in zip(labels, friends, minutes):
    plt.annotate(label, xy=(friend_ct, minute_ct), xytext = (5,-5),textcoords='offset points')
```

```python
plt.title('Daily Minutes vs Number of Friends')
```

```python
plt.xlabel('Number of Friends')
plt.ylabel('Daily Minutes Spent on this Site')
plt.show()
```

```python
test_1_grades = [99,90,85,97,80]
test_2_grades = [100,85,60,90,70]

plt.scatter(test_1_grades, test_2_grades)
plt.title('Axes aren\'t comparable')
plt.xlabel('Test 1 Grade')
plt.ylabel('Test 2 Grade')
plt.show()
```

```python
## Adding plt.axis to make axes comparable
test_1_grades = [99,90,85,97,80]
test_2_grades = [100,85,60,90,70]

plt.scatter(test_1_grades, test_2_grades)
plt.title('Axes are now comparable')
plt.xlabel('Test 1 Grade')
plt.ylabel('Test 2 Grade')
plt.axis('equal') ## Notice plt.axis 
plt.show()
```
